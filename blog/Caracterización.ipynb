```python
drive_path = '/content/gdrive/Shareddrives/TAE/trabajo3/datos/'
df = pd.read_csv(drive_path + 'final.csv', sep = ",", encoding='utf-8')
```

```python
cols_to_graph = ['Con muertos', 'Atropello', 'Ciclo Ruta', 'Glorieta',
       'Interseccion', 'Lote o Predio', 'Paso Elevado', 'Paso Inferior',
       'Paso a Nivel', 'Pontón', 'Puente', 'Tramo de via', 'Tunel',
       'Via peatonal']
x = np.arange(len(cols_to_graph))

nrow = 3; ncol = 2;
fig, axs = plt.subplots(nrows=nrow, ncols=ncol, figsize=(10, 10), constrained_layout = True)

for ax, cluster in zip(axs.reshape(-1), range(6)): 
  datos = {'cluster': [], 'categoria': [], 'cantidad': []}
  df_clust = df[df['cluster'] == cluster]
  

  # TODO: Nombre de gráfica y labels
  for col in cols_to_graph:
    cantidad = df_clust[col].mean()

    datos['cluster'].append(cluster)
    datos['categoria'].append(col)
    datos['cantidad'].append(cantidad)

  sns.barplot(data=datos,
              x='categoria',
              y='cantidad',
              ax=ax,
              order=cols_to_graph,
              orient='v')
  
  ax.set_title(f'carreras cluster: {cluster}')
  ax.set_ylabel('cantidad')
  ax.set_xticks(x)
  ax.set_xticklabels(cols_to_graph, rotation=90)
```

De estas gráficas se puede ver que:

- Los accidentes en tramo de vía son con gran diferencia los de mayor ocurrencia y los accidentes con muertos o en ciclorruta los de menor.
